```python
#Lets import the machine learning models
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import make_pipeline

#This is a classification problem so lets import classification algorithms
from sklearn.tree import DecisionTreeClassifier,ExtraTreeClassifier
#Ensemble techniques
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier,AdaBoostClassifier
#Boosting technique
from xgboost import XGBClassifier
# Logistic regression technique
from sklearn.linear_model import LogisticRegression
#Support vector classifier
from sklearn.svm import SVC
#Kneighborsclassifiers
from sklearn.neighbors import KNeighborsClassifier

#Gaussion process
from sklearn.gaussian_process import GaussianProcessClassifier

#NaiveBayes
from sklearn.naive_bayes import BernoulliNB, GaussianNB

#Evaluation metrics
from sklearn.metrics import accuracy_score, auc, classification_report, roc_auc_score, roc_curve
```

```python
#Lets split the data into train and test

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dtc = DecisionTreeClassifier()
etc = ExtraTreeClassifier()
rfc = RandomForestClassifier()
gbc = GradientBoostingClassifier()
adc = AdaBoostClassifier()
xbg = XGBClassifier()
lg = LogisticRegression()
sv = SVC()
knn = KNeighborsClassifier()
gp  = GaussianProcessClassifier()
bnb = BernoulliNB()
gnb = GaussianNB()

```

```python
algorithms = [dtc,etc,rfc,gbc,adc,xbg,lg,sv,knn,gp,bnb,gnb]
```

```python
def algorithms_list(algorithms):
    for name in algorithms:
        pipeline = make_pipeline(StandardScaler(), name)
        pipeline.fit(x_train,y_train)
        test_predict = pipeline.predict(x_test)
        print(f'Accuracy of the {name} is {accuracy_score(test_predict,y_test)}')
        print(f'Classification report of {name} is {classification_report(y_test,test_predict)}')
        print('*'*30)
```

```python
algorithms_list(algorithms)
```

```python
#The best algorithms are RandomForestClassifier(), GradientBoostingClassifier() and  XGBClassifier()
```

```python
#Lets see the curve 
####################Without  Scaling the input features#####################3

test_prediction = rfc.predict(x_test)
fpr,tpr,thresholds = roc_curve(y_test,test_prediction)
roc_auc = auc(fpr,tpr)
print(roc_auc)
#Plot ROC
plt.title('Receiver operating characteristic')
plt.plot(fpr, tpr,'b',label = 'AUC = %0.2f'% roc_auc)
plt.legend(loc=0)
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.0])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
