```python
#If we see the stop_time and stop duration both are same only specifically stop_time is more specific then stop duration
#Lets deop stop duration column since we already has stop_time
police_df.drop(['stop_duration'], axis=1, inplace=True)
police_df.head(2)
```

```python
police_df['search_conducted'].replace({False:0,
                                   True:1}, inplace=True)

police_df.head(4)
```

```python
police_df['is_arrested'].replace({False:0,
                                   True:1}, inplace=True)
```

```python
police_df.isnull().sum()
```

```python
#Country column is complete null so just delete it
police_df.drop(['country_name'], inplace=True, axis=1)
```

```python
#Lets fill the is arrested column with Zero since 4060 columns are null and in the input boolean value was given
police_df['is_arrested'].fillna(0,axis=0, inplace=True)
```

```python
#Seeing the above since stop_outcome is null we assume that they are not arrested and in the data we have 
#No action value so we will just fill it with no action buddy

police_df['stop_outcome'].fillna('No Action', axis=0, inplace=True)
police_df.head()
```

```python
police_df['search_type'].value_counts()
```

```python
#Just drop the column search type since most of the data in that column is empty :)
police_df.drop(['search_type'], axis=1, inplace=True)
police_df.head(2)
```

```python
police_df['stop_outcome'].value_counts()
```

```python
#one hot encoding
police_df_stop_outcome = pd.get_dummies(police_df['stop_outcome'],prefix='s_o',drop_first=True)
police_df_stop_outcome
```

```python
#Lets concinate the data frame and drop the stop_outcome categorical data

new_police_df = pd.concat([police_df,police_df_stop_outcome],axis=1)
new_police_df.drop(['stop_outcome'],axis=1, inplace=True)
```

```python
new_police_df.head(4)
```

```python
new_police_df['is_arrested'].astype(int)
```

```python
new_police_df.isnull().sum()
```

```python
#new_police_df['violation_raw'].value_counts()
new_police_df['violation_raw'].fillna('Other Traffic Violation', axis=0,inplace=True)
```

```python
#new_police_df['violation_raw'].value_counts()
#Lets do one hot encoding to the column
new_violation = pd.get_dummies(new_police_df['violation_raw'], prefix='voi',drop_first=True)
```

```python
#new_police_df['driver_race'].value_counts()
#Let's fill the driver_race with others since only 4000 columns are present
new_police_df['driver_race'].fillna('Other', axis=0,inplace=True)
```

```python
#Lets do the one hot encoding buddy since wee got the driver race as well
new_race = pd.get_dummies(new_police_df['driver_race'], prefix='race', drop_first=True)
```

```python
new_police_df['driver_age'].mean()
#Mean is 34.14 that is average driver age is 34
```

```python
#Lets fill the age since a continuous numeric data
new_police_df['driver_age'].fillna(new_police_df['driver_age'].mean(), axis=0, inplace=True)
```

```python
#Driver age raw column is not required since we already got the driver age
new_police_df.drop(['driver_age_raw'], axis=1, inplace=True)
new_police_df.head(3)
```

```python
new_police_df[new_police_df['driver_gender'].isnull()]
#Lets replace with mode most repeating value since male is repeate more time lets do it
```

```python
#new_police_df['driver_gender'].mode()
#It's categorical data so lets fill with M and then replace with 0 and 1

new_police_df['driver_gender'].fillna('M', axis=0, inplace=True)
new_police_df['driver_gender'].replace({'M':1,
                                       'F':0}, inplace=True)
new_police_df.head(3)

```

```python
#Til now just handeled the missing values Offf.
#Lets do the time series now 

new_police_df['day']=pd.to_datetime(new_police_df['stop_date'], format='%m/%d/%Y').dt.day
new_police_df['month']=pd.to_datetime(new_police_df['stop_date'], format='%m/%d/%Y').dt.month
new_police_df['year']=pd.to_datetime(new_police_df['stop_date'], format='%m/%d/%Y').dt.year
```

```python
#new_police_df.head()
new_police_df.drop(['stop_date'], axis=1, inplace=True)
```

```python
new_police_df.head()
```

```python
new_police_df[['hour','minute']]=new_police_df['stop_time'].str.split(':').apply(pd.Series)
#Finally splitted the data into hours and minutes
```

```python
#Dropping the stop_time columns since we have already split it perfectly
new_police_df.drop(['stop_time'], axis=1, inplace=True)

#Lets drop violation raw and driver_race as well since we have done one hot encoding i.e get dummies
new_police_df.drop(['violation_raw','driver_race'], axis=1, inplace=True)

#Lets merge the final data frame
final_police_df = pd.concat([new_police_df,new_race,new_violation],axis=1)
final_police_df
```

```python
final_police_df['hour'] = final_police_df['hour'].astype(int)
final_police_df['minute'] = final_police_df['minute'].astype(int)
#Lets convert the data types into int
```

```python
#Building machine learning model
#It's a classification problem

#Pre-processing step
from sklearn.preprocessing import StandardScaler
#Pipeline avoids leakage of data during the ml build
from sklearn.pipeline import make_pipeline,Pipeline


#Spliting the model
from sklearn.model_selection import train_test_split

#Cross validation
from sklearn.model_selection import cross_val_score

#Import stratified k fold validation for binary classification
from sklearn.model_selection import StratifiedKFold

#ML models
#Logistic Regressor model
from sklearn.linear_model import LogisticRegression
#Import support vector classifier
from sklearn.svm import SVC
#Import Random Forest
from sklearn.ensemble import RandomForestClassifier
#import knearest neighbour
from sklearn.neighbors import KNeighborsClassifier

#Hyper parameter tuning for this case I only use Grid search
from sklearn.model_selection import GridSearchCV

#Finally accuracy score
from sklearn.metrics import accuracy_score, f1_score,confusion_matrix,classification_report
```

```python
X = final_police_df.drop('is_arrested', axis=1)
y = final_police_df['is_arrested']
print(X.shape)
print('*'*3)
print(y.shape)
```

```python
#lets split the data
x_train,x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
pipe = make_pipeline(StandardScaler(),LogisticRegression())
pipe.fit(x_train,y_train)
```

```python
x_train_pipe = pipe.predict(x_train)
train_accuracy = accuracy_score(x_train_pipe, y_train)
print(train_accuracy)
```

```python
x_test_pipe = pipe.predict(x_test)
test_accuracy = accuracy_score(x_test_pipe, y_test)
print(test_accuracy)
```

```python
pipe2 = make_pipeline(StandardScaler(), RandomForestClassifier())
pipe2.fit(x_train,y_train)
```

```python
x_train_pipe = pipe2.predict(x_train)
train_accuracy = accuracy_score(x_train_pipe, y_train)
print(f'Training accuracy is {train_accuracy}')
print('*'*30)
x_test_pipe = pipe2.predict(x_test)
test_accuracy = accuracy_score(x_test_pipe, y_test)
print(f'Test Accuracy is {test_accuracy}')
```

```python
#Lets do the shuffle with strafiedkfold and do cross val score
rfc = RandomForestClassifier()
skf = StratifiedKFold(n_splits=3)
cross_val_rfc = cross_val_score(estimator = rfc,X=X, y=y,cv=skf, verbose=2)
print(cross_val_rfc)
print(np.mean(cross_val_rfc))
```

We can see 99.7 % looks very good

```python
#Lets do the same thing for Logistic Regression lets see what we can get

LR = LogisticRegression()
skf = StratifiedKFold(n_splits=3)
cross_val_rfc = cross_val_score(estimator = LR,X=X, y=y,cv=skf, verbose=2)
print(cross_val_rfc)
print(np.mean(cross_val_rfc))
```

logistic regression provided 97% accuracy so lets use Random forest
